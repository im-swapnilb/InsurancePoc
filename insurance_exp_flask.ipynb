```python
df = pd.read_csv("insurance.csv")
df.head()
```

```python
plt.scatter(df.age, df.charges, marker = '*', color = 'green')
```

```python
df.shape
```

```python
sex = np.zeros(len(df))
sex[df['sex']== 'male'] = 1
sex[df['sex']== 'female'] = 0
df['sex'] = sex
```

```python
smoker = np.zeros(len(df))
smoker[df['smoker']== 'yes'] = 1
smoker[df['smoker']== 'no'] = 0
df['smoker'] = smoker

df.head(20)
```

```python
# Checking if there are any missing values in the data.
df.isnull().sum()
```

```python
X = df.drop(['charges', 'region'], axis=1)
y = df['charges']
```

```python
# Sperate train and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=5)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
y
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
linear_predictions = model.predict(X_test)
linear_predictions_train = model.predict(X_train)
```

```python
model.score(X_test, linear_predictions)
```

```python
np.sqrt(mean_squared_error(y_test, linear_predictions))
```

```python
pickle_out = open('insurance_exp_flask.pkl','wb')
pickle.dump(model, pickle_out)
loaded_model = pickle.load(open('insurance_exp_flask.pkl','rb'))
result = loaded_model.score(X_test, linear_predictions)
print(result)
```
